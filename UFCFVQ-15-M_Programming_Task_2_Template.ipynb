### Requirement FR9 - Read CSV data from two files and merge it into a single Data Frame 

```python
#import pandas library
import pandas as pd

# #identifying csv files to be read
# csv_files = ["task2a.csv", "task2b.csv"]

# #create list of dataframes 
# dataframes = []

# #loop csv files into dataframes
# for file in csv_files:
#     dataframes.merge(pd.read_csv(file))

# df = pd.concat(dataframes)

# df.info()


#read first csv data
task2a = pd.read_csv("task2a.csv")

#read second csv data
task2b = pd.read_csv("task2b.csv")



#merge both csv files into one
df = pd.merge(task2a, task2b, on = "region")
df
# df.info()
```

### Requirement FR11 - Investigate the effects of household income on levels of bias crimes

```python
# import matplotlib.pyplot as plt

# %matplotlib inline

# # plt.style.available
# plt.style.use('grayscale')

# df.plot(x = 'median_household_income', y = "avg_biascrimes_per_100k", kind = "scatter")


#import seaborn and matplotlib
import seaborn as sns
import matplotlib.pyplot as plt

#plot graph declaring axis and title
sns.scatterplot(
    data = df,
    x="median_household_income", 
    y="avg_biascrimes_per_100k",)
plt.title('Effect of household income on level of bias crime')
plt.show()

```

From the plot above, it is obvious that median household income had no significant effect on level of bias crimes.Generally,  it would be expected that regions with lesser household income should have higher biased crimes, but that wasn't the case. Regions with higher bias crimes didnt have significant lower bias crimes either

### Requirement FR12 - Test the hypothesis that household income has no effect on levels of bias crime

```python
# add code here
import scipy.stats

r, p_value = scipy.stats.pearsonr(df['median_household_income'], df['avg_biascrimes_per_100k'])

#print values of correlation and p-value to 4 decimal places
print(f'The correlation coefficient r={r:.4f}, and p-value={p_value:.4f}\n')
```

the co-efficient of correlation is closer to 0, so we can deduce that there's no correlation between household income and level of bias crime. Meaning, household income has no effect on levels of bias crimes.
